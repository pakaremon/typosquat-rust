```python
import json
import pandas as pd
import plotly.express as px
```

```python
with open("/unique_dns and packages.json", "r") as json_file:
  data = json.load(json_file)
```

```python

data_for_df = []

for item in data:
  data_for_df.append({'Hostname': item['Hostname'], 'Name_Count': len(item['Names'])})

df = pd.DataFrame(data_for_df)
df
```

```python

fig = px.pie(df, values='Name_Count', names='Hostname', title='Name Count per Hostname')
fig.show()
```
